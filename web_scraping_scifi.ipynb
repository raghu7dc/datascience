```python
#import necessary libraries
import requests
from bs4 import BeautifulSoup as bs
```

```python
#load the web page
url = 'https://en.wikipedia.org/wiki/List_of_highest-grossing_science_fiction_films#Highest-grossing_sci-fi_films'
r = requests.get(url)
```

```python
#convert to beautiful soup object
soup=bs(r.content)
```

```python
#print our html
print(soup.prettify())
```

```python
first_header=soup.find_all('h2')
print(first_header)
```

```python
#pass in a list
print(soup.find(['h1','h2']))
headers= soup.find_all(['h1','h2'])
print(headers)
```

```python
#passing attributes to find
paragraph=soup.find_all('p',attrs={'id':"cite_ref-Bialik_(2007)_58-0"})
paragraph
```

```python
#nest find/find_all
body=soup.find('body')
body
div=body.find('div')
div
```

```python
header=div.find('h1')
header
```

```python
import re
soup.find_all('p',string=re.compile('science fiction'))
headers==soup.find_all('h2',attrs={'id':'Highest-grossing_sci-fi_films'})
headers
```

```python
#highest grossing films of all time
table=soup.find_all('table')
#wikitable sortable gives the main content table
content_table=soup.find('table',{'class':'wikitable sortable'}) #find rather than find_all ,since only one table has wikitable sortable
content_table

rows=content_table.find_all('tr')
for row in rows:
  print(row.get_text())
```

```python
#highest grossing films adjusted for inflation
table=soup.find_all('table')
#wikitable plainrowheaders gives the another content table
content_table=soup.find('table',{'class':'wikitable plainrowheaders'}) 
content_table

rows=content_table.find_all('tr')
for row in rows:
  print(row.get_text())
```

```python
#highest grossing films adjusted for inflation and timeline of highest grossing film
table=soup.find_all('table')
#wikitable plainrowheaders gives the another content table
content_table=soup.find_all('table',{'class':'wikitable plainrowheaders'}) 
content_table

for table in content_table:
  rows=table.find_all('tr')
  for row in rows:
    print(row.get_text())
```

```python
#highest grossing film-franchises and its films
table=soup.find_all('table')
#wikitable plainrowheaders gives the another content table
content_table=soup.find_all('table',{'class':'wikitable collapsible collapsed'}) 
content_table


for table in content_table:
  rows=table.find_all('tr')
  for row in rows:
    print(row.get_text())
```

```python
#getting a specific moviename from table
table=soup.find('table',{'class':'wikitable sortable'})
rows=table.find_all('a',title=re.compile('Avatar'))
print(rows)
for row in rows:
  print(row.get_text())
```
